We load the data.

```python
calendar_events = pd.read_csv('calendar_events.csv')
calendar = pd.read_csv('calendar.csv')
items_weekly_sell_prices = pd.read_csv('items_weekly_sell_prices.csv')
sales_test = pd.read_csv('sales_test.csv')
sales_train = pd.read_csv('sales_train.csv')
```

We now look at calendar

```python
calendar.head(10)
```

```python
calendar.tail(10)
```

```python
calendar.dtypes
```

```python
calendar['date'] = pd.to_datetime(calendar['date'])
```

```python
calendar.describe(include=['object', 'category',"datetime64","int64"])
```

```python
calendar['d'] = calendar['d'].astype('str')
```

```python
calendar['year'] = (calendar['wm_yr_wk'] // 100).astype(int) + 1900  # Extract and adjust year
calendar['week'] = (calendar['wm_yr_wk'] % 100).astype(int)
```

```python
calendar.describe(include=['object', 'category',"datetime64","int64"])
```

```python
calendar.tail(10)
```

```python
calendar['year'] = calendar['date'].dt.year
calendar['week'] = calendar['date'].dt.isocalendar().week
```

```python
calendar.dtypes
```

```python
len(calendar)
```

```python
calendar.head(10)
```

```python
calendar.isnull().sum()
```

We join calendar and calendar_events together

```python
calendar_merged = pd.merge(calendar, calendar_events, how='left', on='date')
```

```python
calendar_merged
```

```python
calendar_merged.describe(include=['object', 'category',"datetime64","int64"])
```

```python
calendar_merged.isnull().sum()
```

```python
for col in calendar_merged.select_dtypes(['category']).columns:
    calendar_merged[col] = calendar_merged[col].cat.add_categories("None")
calendar_merged.fillna("None")
```

We now look at items_weekly_sell_prices

```python
items_weekly_sell_prices
```

```python
items_weekly_sell_prices.dtypes
```

```python
items_weekly_sell_prices['store_id'] = items_weekly_sell_prices['store_id'].astype('str')
items_weekly_sell_prices['item_id'] = items_weekly_sell_prices['item_id'].astype('str')
```

```python
items_weekly_sell_prices.isnull().sum()
```

we now look at sales_train

```python
sales_train
```

```python
sales_train_melt = pd.melt(sales_train, 
                  id_vars = ['id','item_id','dept_id','cat_id','store_id','state_id'], 
                  var_name = 'd', 
                  value_name = "sales")
```

```python
sales_train_melt
```

```python
sales_train_processed = pd.merge(sales_train_melt, calendar_merged, how='left', on=['d'])
```

```python
sales_train_processed = pd.merge(sales_train_processed, items_weekly_sell_prices, how='left', on=['store_id', 'item_id', 'wm_yr_wk'])
```

```python
sales_train_processed.isnull().sum()
```

```python
sales_train_processed.head(10)
```

```python
sales_train_final = sales_train_processed.drop(['id', 'd',"wm_yr_wk"], axis=1)
```

```python
sales_train_final.head(6)
```

```python
sales_train_final["revenue"] = sales_train_final["sales"]*sales_train_final["sell_price"]
```

```python
sales_train_final.head(6)
```

```python
sales_train_final['event_name'] = sales_train_final['event_name'].fillna("None")
sales_train_final['event_type'] = sales_train_final['event_type'].fillna("None")
sales_train_final['revenue'] = sales_train_final['revenue'].fillna(0)
```

```python
sales_train_final.describe()
```

```python
sales_train_final.dtypes
```

```python
sales_train_final.head(6)
```

```python
sales_train_final.to_csv('/Users/alexanderschou/Desktop/Projects/adv_mla_2024/AT2_experimentation/data/processed/sales_train_processed.csv', index=False)
```
